```python
import matplotlib.pyplot as plt

from skimage.feature import hog
from skimage import data, exposure


image = data.brick()
```

```python
print(image)
```

```python
from skimage.feature import local_binary_pattern
lbp_image=local_binary_pattern(image,8,2,method='uniform')
print(lbp_image)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)
ax1.axis('off')
ax1.imshow(image)
ax1.set_title('Input image')

# Rescale histogram for better display
lbp_image_rescaled = exposure.rescale_intensity(lbp_image, in_range=(0, 10))

ax2.axis('off')
ax2.imshow(lbp_image_rescaled, cmap=plt.cm.gray)
ax2.set_title('Local binary patterns')
plt.show()
```
